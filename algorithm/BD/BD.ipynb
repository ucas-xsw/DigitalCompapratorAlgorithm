```python
import numpy as np
import matplotlib.pyplot as plt

def objective(x):
    return (x[0] - 1) ** 2 + (x[1] - 3) ** 2

class Node:
    def __init__(self, x, bounds):
        self.x = x
        self.bounds = bounds
        self.value = None

def branch_and_bound(f, bounds, max_iter=100):
    nodes = [Node(None, bounds)]
    opt_val = None
    opt_x = None
    iter_count = 0
    opt_values = []
    while nodes:
        node = nodes.pop(0)
        if iter_count >= max_iter:
            break
        if opt_val is not None and node.bounds[0][0] > opt_val:
            continue
        if node.x is None:
            for i in range(len(bounds)):
                for j in range(bounds[i][0], bounds[i][1]+1):
                    x = [None] * len(bounds)
                    x[i] = j
                    nodes.append(Node(x, bounds))
        else:
            if node.value is None and node.x is not None and node.bounds is not None:
                node.value = f(node.x)
                opt_values.append(node.value)
                if opt_val is None or node.value < opt_val:
                    opt_val = node.value
                    opt_x = node.x
            else:
                if opt_val is None or node.value < opt_val:
                    for i in range(len(bounds)):
                        if node.x[i] is None:
                            for j in range(node.bounds[i][0], node.bounds[i][1]+1):
                                x = node.x.copy()
                                x[i] = j
                                nodes.append(Node(x, node.bounds))
    plt.plot(opt_values)
    plt.xlabel('Iterations')
    plt.ylabel('Optimal Value')
    plt.show()
    return opt_val, opt_x


bounds = [(-10, 10), (-10, 10)]
opt_val, opt_x = branch_and_bound(objective, bounds, max_iter=100)
print('Optimal value:', opt_val)
print('Optimal solution:', opt_x)
```
